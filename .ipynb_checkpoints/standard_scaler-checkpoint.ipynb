```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import sklearn as sk 
import seaborn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r'Social_Network_Ads.csv')
df.head()
```

```python
df.drop('User ID',inplace=True,axis=1)
df.head()
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
df
```

```python
df
```

```python
#train trest split
x= df.drop('Purchased',axis=1)
y= df['Purchased']
x_test ,x_train ,y_test, y_train =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ss = StandardScaler()
x_train_s = ss.fit_transform(x_train)
x_test_s = ss.fit_transform(x_test)
x_train_s
```

```python
fig,(x1,x2) = plt.subplots(1,2,figsize=(15,4))
x1.scatter(x_train['Age'],x_train['EstimatedSalary'])
plt.title("before")
plt.xlabel("Age")
plt.ylabel('Salary')
x2.scatter(x_train_s['Age'],x_train_s['EstimatedSalary'],color='red')
plt.title('after scaling')
plt.xlabel("Age")
plt.ylabel('Salary')
plt.show()
```
